```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mlxtend.evaluate import bias_variance_decomp
from sklearn.neural_network import MLPClassifier
%matplotlib inline 
```

```python
train=pd.read_csv("./train.csv")
train.head()
```

```python
test=pd.read_csv("./test.csv")
test.head()
```

```python
train_=train.drop(["Id"], axis=1)
train_.head()
```

```python
# let test and train's column name be the same
columns_name=list(train_.columns)
test.columns=columns_name
test.head()
```

```python
X_train=train_.drop(["Category"],axis=1)
y_train=train_["Category"]
X_test=test.drop(["Category"],axis=1)
y_test=test["Category"]
#print(X_train)
#print(y_train)
#print(X_test)
#print(y_test)
```

```python
hidden_layers=[1,2,3]
hidden_nodes=[50,200,784]
optimizers=['adam','sgd']
learning_rate_init=None
for i in range(2):
    for j in range(3):
        if optimizers[i]=='sgd':
            learning_rate_init=0.0001
        else:
            learning_rate_init=0.001
        clf = MLPClassifier(hidden_layer_sizes=(hidden_nodes[j]),
                            solver=optimizers[i],early_stopping=True,
                            learning_rate_init=learning_rate_init,
                            max_iter=100).fit(X_train, y_train)
        print("# of hidden_layers=1, hidden_nodes={}, optimizer={}".format(hidden_nodes[j],optimizers[i]))
        print(f"Train Accuracy : {clf.score(X_train, y_train)}")
        print(f"Test  Accuracy : {clf.score(X_test, y_test)}")
        clf = MLPClassifier(hidden_layer_sizes=(hidden_nodes[j],hidden_nodes[j]),
                            solver=optimizers[i],
                            early_stopping=True,
                            learning_rate_init=learning_rate_init,
                            max_iter=100).fit(X_train, y_train)
        print("# of hidden_layers=2, hidden_nodes={}, optimizer={}".format(hidden_nodes[j],optimizers[i]))
        print(f"Train Accuracy : {clf.score(X_train, y_train)}")
        print(f"Test  Accuracy : {clf.score(X_test, y_test)}")
        clf = MLPClassifier(hidden_layer_sizes=(hidden_nodes[j],hidden_nodes[j],hidden_nodes[j]),
                            solver=optimizers[i],
                            early_stopping=True,
                            learning_rate_init=learning_rate_init,
                            max_iter=100).fit(X_train, y_train)
        print("# of hidden_layers=3, hidden_nodes={}, optimizer={}".format(hidden_nodes[j],optimizers[i]))
        print(f"Train Accuracy : {clf.score(X_train, y_train)}")
        print(f"Test  Accuracy : {clf.score(X_test, y_test)}")
```
